```python
# 17. Creating a new column in the existing dataframe as “amount_in_usd” and convert all the non-USD currencies in USD and store them in the same column. 
# takes about a minute
exchange_rates = {'EUR': 1.08,'GBP': 1.24, 'CAD': 0.74, 'PLN': 0.24,'AED': 0.27, 'CHF': 1.10, 'RON': 0.22, 'CZK': 0.045, 'HUF': 0.0029,'BHD': 2.65, 'SAR': 0.27, 'QAR': 0.27, 'KWD': 3.25, 'SEK': 0.093}
def convertUSD(row):
    cur = row['ORDER_CURRENCY']
    ex_rate = exchange_rates.get(cur, 1)
    order_amount = pd.to_numeric(row['ORDER_AMOUNT'])
    return order_amount * ex_rate

df2['amount_in_usd'] = df2.apply(convertUSD, axis=1)
```

<font size="5">2. Performing Mean encoding on all the categorical columns 
</font>

```python
# label encoding PURCHASE_ORDER_TYPE
df5 = df4.copy()

print(df5.PURCHASE_ORDER_TYPE.isna().sum())
df5 = df5.dropna(subset=['PURCHASE_ORDER_TYPE'])

median_credit_status = df5["CREDIT_STATUS"].median()
df5["CREDIT_STATUS"].fillna(median_credit_status, inplace=True)

df5
```

<font size='5'>Here I have used Mean encoding is used to encode categorical variables by replacing each category with the mean target value. Mean encoding captures the relationship between variables and the target, providing more informative representations. Mean encoding is particularly beneficial for large datasets with a large number of variables as it avoids creating high-dimensional sparse matrices and helps mitigate memory and computational challenges.</font>

```python
 columns_to_mean_encode = ['DISTRIBUTION_CHANNEL', 'DIVISION', 'PURCHASE_ORDER_TYPE', 'CREDIT_CONTROL_AREA', 'ORDER_CURRENCY','COMPANY_CODE']

# Define the noise factor(will be used to prevent multiple unique values being assigned same labels because of same mean values)
noise_factor = 0.001

# Perform mean encoding
for column in columns_to_mean_encode:
    # Calculate the mean of 'amount_in_usd' for each category in the column
    mean_encoded = df5.groupby(column)['amount_in_usd'].mean()
    
    # Add noise to the mean values
    mean_encoded += np.random.randn(len(mean_encoded)) * noise_factor
    
    mean_encoded = mean_encoded.round().astype(int)
    
    # Map the mean values to the corresponding categories in the column
    df5[column] = df5[column].map(mean_encoded)

# Dataset after label encoding:
df5
```

<font size='5'>This function calculates the differences in sales for specific days by shifting the sales values for each customer and calculating the differences between the current day and the previous days. The function takes into account the specified lag and starting day, allowing for flexible datetime manipulation and analysis of sales data.</font>

```python
def difference_in_days(melt, lags, ffday, customer_id_col, create_date_col, net_amount_col):
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Sales'] = melt.groupby([customer_id_col])[net_amount_col].shift(i)
    
    melt = melt.reset_index(drop = True)
    
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Diff']  = melt.groupby([customer_id_col])['Last-'+str(i)+'day_Sales'].diff()
    melt = melt.fillna(0)
    return melt
```
